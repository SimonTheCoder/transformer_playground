```python
from transformers import GPTNeoForCausalLM, GPT2Tokenizer

model = GPTNeoForCausalLM.from_pretrained("EleutherAI/gpt-neo-1.3B")
tokenizer = GPT2Tokenizer.from_pretrained("EleutherAI/gpt-neo-1.3B")

# model = GPTNeoForCausalLM.from_pretrained("EleutherAI/gpt-neo-2.7B")
# tokenizer = GPT2Tokenizer.from_pretrained("EleutherAI/gpt-neo-2.7B")

```

```python
import torch
# Set the model to use the GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
# device = "cpu"
model.to(device)
```

```python
# prompt = (
#     "In a shocking finding, scientists discovered a herd of unicorns living in a remote, "
#     "previously unexplored valley, in the Andes Mountains. Even more surprising to the "
#     "researchers was the fact that the unicorns spoke perfect English."
# )
# prompt = "a python script which generate a random number between 0 to 42:"
prompt = "calculate: 1+1="

input_ids = tokenizer(prompt, return_tensors="pt").input_ids

gen_tokens = model.generate(
    input_ids.to(device),
    do_sample=True,
    temperature=0.5,
    max_length=100,
    repetition_penalty = 1.0,
)
print()
gen_text_list = tokenizer.batch_decode(gen_tokens)
gen_text=gen_text_list[0]
print(prompt)
print(gen_text)
print("="*10)
# print(gen_text_list)
```

```python
# input_text = "# this funcion return a random number between 0 to 42. \ndef random_int():"
input_text = "#get a user input, try to convert it to a int and return it. If failed return 0. \ndef get_userinput():"

# Encode the input text
input_ids = tokenizer.encode(input_text, return_tensors="pt")

# Generate text
output = model.generate(input_ids.to(device),
    max_length = 1500,
    temperature = 0.1,
    repetition_penalty = 1.0,
    do_sample = True
)

# Decode the output text
output_text = tokenizer.decode(output[0], skip_special_tokens=True)

print(output_text)
```
